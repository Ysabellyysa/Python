```python
import pandas as pd #importando o pandas
```

```python
sal = pd.read_csv('Salaries.csv') #Ler o arquivo salaries 
```

```python
sal.head() #verifica o Head do DataFrame
```

```python
sal.info() #Quantidade de entradas 
```

```python
sal['BasePay'].mean() #Média do "Salário Base" 
```

```python
sal['OvertimePay'].max() #maior quantidade no conjunto de dados 
```

```python
sal[sal['EmployeeName']=='JOSEPH DRISCOLL']['JobTitle'] #Cargo do Joseph Driscoll
```

```python
sal[sal['EmployeeName']=='JOSEPH DRISCOLL']['TotalPayBenefits'] #Sálario do Joseph Driscoll
```

```python
sal[sal['TotalPayBenefits']==sal['TotalPayBenefits'].max()]
```

```python
sal[sal['TotalPayBenefits']==sal['TotalPayBenefits'].min()] #Pessoa paga com o salário mais baixo
```

```python
sal.groupby('Year')['BasePay'].mean() #Média salarial de todos os funcionarios 
```

```python
sal['JobTitle'].nunique() #titulos de trabalho
```

```python
sal['JobTitle'].value_counts().head(5) #5 principais empregos
```

```python
sum(sal[sal['Year']==2013]['JobTitle'].value_counts() == 1)
```

```python
def chief_string(title):
    if 'chief' in title.lower():
        return True
    else:
        return False
```

```python
sum(sal['JobTitle'].apply(lambda x: chief_string(x)))
```

```python
sal['title_len'] = sal['JobTitle'].apply(len)
```

```python
sal[['title_len','TotalPayBenefits']].corr()
```
