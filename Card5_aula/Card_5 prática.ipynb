```python
#Media e Mediana 
gastos_comuns = np.random.normal(20, 5, 1000) #Gera os valores
print(f'Média: {np.mean(gastos_comuns):.2f}') #calcula e gera o valor médio
print(f'Mediana: {np.median(gastos_comuns):.2f}') #Acha e gera o valor do meio

gastos_erros = np.append(gastos_comuns, [50000])#gera os valores
print(f'Média: {np.mean(gastos_erros):.2f}') #calcula e gera o valor médio
print(f'Mediana: {np.median(gastos_erros):.2f}') #Acha e gera o valor do meio
```

```python
#Probabilidade Condicional
totals = {'Cafe_comum': 500, 'Cafe_especial': 300} #total de pedidos de café
sobremesas = {'Cafe_comum': 100, 'Cafe_especial': 210} #total de pedidos de café com doces

P_condicional = float(sobremesas['Cafe_especial']) / float(totals['Cafe_especial'])
#calculo de probabilidade condicional
print(f'Sobremesa | Café Especial: {P_condicional:.2%}') #imprime o resultado
```

```python
#Probabilidade Geral
total_pedidos = 800 #500 comuns + 300 especias - total de pedidos
total_sobremesas = 310 # 100 comuns e 210 comuns - total de pedidos


PE = float(total_sobremesas) / float(total_pedidos)
#Probabilidade de qualquer cliente pedir sobremesa
print(f"Sobremesa Geral: {PE:.2%}") #imprime o resultado
```

```python
#Probabilidades 
from numpy import random #importando bibliotecas
random.seed(0)

# armazenar totais de clientes e quem tomou café por idade
totals = {20:0, 30:0, 40:0, 50:0, 60:0, 70:0}
cafe_bebidas = {20:0, 30:0, 40:0, 50:0, 60:0, 70:0}
total_cafe = 0

for _ in range(100000):
    # Escolhe uma faixa de idade aleatória
    ageDecade = random.choice([20, 30, 40, 50, 60, 70])
    
    # Define a chance de tomar café baseada na idade mais velho = mais chance
    # Aqui cria uma lógica onde a probabilidade aumenta com a idade
    chance_por_idade = {20: 0.3, 30: 0.4, 40: 0.6, 50: 0.7, 60: 0.8, 70: 0.9}
    
    purchaseProbability = chance_por_idade[ageDecade]
    
    totals[ageDecade] += 1
    
    if (random.random() < purchaseProbability):
        total_cafe += 1
        cafe_bebidas[ageDecade] += 1

# Exemplo de cálculo de probabilidade condicional
PEF_20 = float(cafe_bebidas[20]) / float(totals[20])
print(f"Café | 20 anos: {PEF_20:.4f}")

PEF_70 = float(cafe_bebidas[70]) / float(totals[70])
print(f"Café | 70 anos: {PEF_70:.4f}")

# Probabilidade geral
PE = float(total_cafe) / 100000.0
print(f"Café Geral: {PE:.4f}")
```

```python
import matplotlib.pyplot as plt

# Calculando as probabilidades para cada idade para o gráfico
idades = list(totals.keys())
probs = [cafe_bebidas[age] / totals[age] for age in idades]

plt.bar(idades, probs, width=5) #criação do gráfico de barras
plt.title("Probabilidade de Tomar Café por Idade") #titulo do gráfico
plt.xlabel("Faixa Etária") #eixo x 
plt.ylabel("Probabilidade") #eixo y
plt.show() #gera o gráfico
```
