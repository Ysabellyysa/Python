```python
#importando as bibliotecas necessárias
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
t = np.arange(0.0, 2.0, 0.01) #cria um intervalo 
s = 1 + np.sin(2 * np.pi * t) #calcula o valor da formala para cada ponto 

#criação do gráfico 
fig, ax = plt.subplots()
ax.plot(t,s)

ax.set(xlabel='Tempo', ylabel='Voltagem', 
      title='Gráfico de Seno')
#configuração dos rótulos do eixos e do título do gráfico
ax.grid()
#adiciona uma grelha ao fundo para facilitar a leitura dos valores 

plt.show() #Gera o gráfico
```
