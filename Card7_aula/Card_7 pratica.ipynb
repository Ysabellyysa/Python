```python
#importando as bibliotecas necessárias
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
# Definindo o horário 
horas = np.linspace(8, 20, 100) 

fluxo_clientes = (np.exp(-(horas - 9)**2 / 2) +  # calcula o exponencial do pico das 9
                  np.exp(-(horas - 14)**2 / 5) + # calcula o expotencial do pico das 14
                  0.2 * np.random.rand(100))   

# Multiplicamos para ter números inteiros de clientes
fluxo_clientes = fluxo_clientes * 40 

#Criação do gráfico
fig, ax = plt.subplots(figsize=(10, 5))

# Criando um gráfico de preenchimento
ax.plot(horas, fluxo_clientes, color="black")

ax.set_title('Fluxo de Clientes') #Título do gráfico
ax.set_xlabel('Horário de Funcionamento') #Eixo x
ax.set_ylabel('Número de Clientes') #Eixo y

# Ajustando os marcadores do eixo X para mostrar as horas certas
ax.set_xticks(range(8, 21))

ax.grid() #adiciona grade no gráfico

plt.show() #Gera o gráfico
```

```python
#Importando as bibliotecas necessárias
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
dados = {'categoria': ['Ficção', 'AutoAjuda', 'Biografia', 'Infantil', 'Técnico'],
        'vendidos': [450, 320, 150, 280, 500]} #dados utilizados no gráfico

df = pd.DataFrame(dados) #Data Frame

df.plot(kind='bar', x='categoria', y='vendidos', color='purple')
#faz o gráfico em versão barra comparando os livros, cor das barras

plt.title('Livros Mais Vendidos') #titulo do gráfico
plt.ylabel('Quantidade de Unidades') #eixo y
plt.xticks(rotation=35) #inclina o texto para não amontoar

plt.show() #gera o código 
```

```python
import matplotlib.pyplot as plt #importando a biblioteca 
```

```python
dias = ['Seg', 'ter', 'qua', 'qui', 'sex']
paginas = [15, 20, 10, 35, 50] #Dados utilizados no gráfico

#Criação de gráfico
fig, ax = plt.subplots()
plt.plot(dias, paginas, color='red', linewidth=2)
#faz o gráfico em linha, linha na cor vermelha

plt.title('Leitura Semanal') #titulo do gráfico
plt.xlabel('Dias da Semana') #eixo x
plt.ylabel('Quantidade de Páginas') #eixo y
plt.grid() #grade do gráfico
plt.show() #Gera o gráfico
```
