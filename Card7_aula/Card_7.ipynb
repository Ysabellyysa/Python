```python
import numpy as np #importa a biblioteca do numpy
import matplotlib.pylab as plt #importa a biblioteca de criação de graficos 
import matplotlib #importação geral da biblioteca
```

```python
#Data for plotting 
t = np.arange(0.0, 2.0, 0.01) #cria um intervalo de 0.0 a 0.2 cm incremento de 0.1
s = 1 + np.sin(2* np.pi * t) #calcula o valor da formala para cada ponto de t usando a função seno

#criação do gr
fig, ax = plt.subplots()
ax.plot(t,s)

ax.set(xlabel='time(s)', ylabel = 'voltagem (mV)',
        title = 'About as simple as it gets, folks')
ax.grid()
```
