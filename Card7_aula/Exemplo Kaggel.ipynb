```python
#Importando as bibliotecas necessárias
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
dados = {'categoria': ['Ficção', 'AutoAjuda', 'Biografia', 'Infantil', 'Técnico'],
        'vendidos': [450, 320, 150, 280, 500]} #dados utilizados no gráfico

df = pd.DataFrame(dados) #Data Frame

df.plot(kind='bar', x='categoria', y='vendidos', color='purple')
#faz o gráfico em versão barra comparando os livros, cor das barras

plt.title('Livros Mais Vendidos') #titulo do gráfico
plt.ylabel('Quantidade de Unidades') #eixo y
plt.xticks(rotation=35) #inclina o texto para não amontoar

plt.show() #gera o código 
```
