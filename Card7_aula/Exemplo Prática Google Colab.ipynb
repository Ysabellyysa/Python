```python
import matplotlib.pyplot as plt
```

```python
dias = ['Seg', 'ter', 'qua', 'qui', 'sex']
paginas = [15, 20, 10, 35, 50] #Dados utilizados no gráfico

#Criação de gráfico
fig, ax = plt.subplots()
plt.plot(dias, paginas, color='red', linewidth=2)
#faz o gráfico em linha, linha na cor vermelha

plt.title('Leitura Semanal') #titulo do gráfico
plt.xlabel('Dias da Semana') #eixo x
plt.ylabel('Quantidade de Páginas') #eixo y
plt.grid() #grade do gráfico
plt.show() #Gera o gráfico
```
