## Carregar primeira página

## Comece a usar o Beautiful Soup para fazer scraping.


find and find_all

```python
first_header = soup.find("h2") #Retorna apenas o primeiro elemento

headers = soup.find_all("h2") #Retorna uma lista com todos os elementos
print(headers)
```

```python
#Insira uma lista de elementos para procurar
first_header = soup.find(["h1", "h2"]) 

headers = soup.find_all(["h1", "h2"]) #find_all captura a estrutura de tópicos do texto
headers
```

```python
#Você pode passar atributos para a função find/find_all.
paragraph = soup.find_all("p", attrs={"id": "paragraph-id"})
paragraph 
```

```python
#Você pode aninhar chamadas find/find_all.
#Busca em cascara
body = soup.find("body") #isola todo o body do site
div = body.find('div') #procura a primeira div dentro do body
header = div.find('h1') #procura o primeiro h1 dentro da div
header 
```

```python
#Podemos pesquisar strings específicas em nossas chamadas find/find_all.
#print(soup.prettify())
import re 

paragraphs = soup.find_all("p", string=re.compile("Some")) #procuta todos os paragrafos
paragraphs

headers = soup.find_all("h2", string=re.compile("(H|h)eader"))
#procuras todos os h2 onde tenha header
headers
```

### Carregando a página 

```python
from bs4 import BeautifulSoup as bs 
#carregando a página content 
r = requests.get("https://keithgalli.github.io/web-scraping/webpage.html")

# Converter beautifull soup para objeto
webpage = bs(r.content)

#print out our html
print(webpage.prettify())
```

### segundo exercício - estrair os dados da tabela 

```python
#l = []
#for tr in table_rows:
#    td = tr.find_all('td')
#    row = [tr.text for tr in td]
#    l.append(row)
#pd.DataFrame(l, columns=['A', 'B', ...])
```

```python
import pandas as pd #importa a bibloteca pandas

table = webpage.select("table.hockey-stats")[0] #localiza a tabela com a classe
columns = table.find("thead").find_all("th") #captura todos os títulos das colunas
columns_names = [c.string for c in columns] #cria a lista em uma unica linha

table_rows = table.find("tbody").find_all("tr")
l = []
for tr in table_rows:
    #para cada linha da tabela busca todas as células de dados
    td = tr.find_all('td')
    row = [str(tr.get_text()).strip() for tr in td] #remove os espaçoes em branco
    l.append(row) #cria uma matriz onde cada item é uma linha
    
df = pd.DataFrame(l, columns=columns_names) 
#pega a lista e os nomes das colunas e monta a tabela
df.loc[df['Team'] != 'Did not play'] #limpeza de dados
```

### Anote todas as curiosidades que usam a palavra "is"

```python
import re

facts = webpage.select("ul.fun-facts is") #seleciona todos os itens da 'li'
facts_with_is = [facts.find_all(string=re.compile("is")) for fact in facts]
#Para cada item da lista ele procura se a palavra 'is' existe dentro 
facts_with_is = [fact.find_parent().get_text() for fact in facts_with_is if fact]
#retorna apenas o nó do texto 
print(facts_with_is) #extrai o texto completo
```

### Baixando imagems

```python
import requests 
from bs4 import BeautifulSoup as bs #importa as bibliotecas

url = "https://keithgalli.github.io//web-scraping/" 
r = requests.get(url+"webpage.html")

webpage = bs(r.content)

images = webpage.select("div.row div.column img") #busca a imagens 
image_url = images[0]['src'] #forma o endereço real onde a imagem fica
full_url = url + image_url

img_data = requests.get(full_url).content #retorna a resposta em byter
with open('image_name.jpg', 'wb') as handler:
    handler.write(img_data)
    
```

### Resolva o desafio misterioso

```python
#mapeia todos os links dentro de uma seção específica
files = webpage.select("div.block a") 
relative_files = [f['href'] for f in files] 

url = "https://keithgalli.github.io//web-scraping/"
for f in relative_files:
    full_url = url + f #monta o endereço completo 
    page = requests.get(full_url)
    bs_page = bs(page.content)
    #passa um dicionario de atributos
    secret_word_element = bs_page.find("p", attrs={"id": "secret-word"})
    secret_word = secret_word_element.string #extrai apenas a palavra secreta da página
    print(secret_word) #imprime a palavra secreta
```
